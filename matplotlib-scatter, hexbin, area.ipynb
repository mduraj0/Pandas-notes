```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
marathon = pd.read_csv("../data/marathon_results_2017.csv", usecols=["Age","M/F","Country","40K"])
marathon["TimeSeconds"] = marathon["40K"].replace('-', 0).apply(lambda x: pd.Timedelta(x).total_seconds())
marathon.head(5)
```

```python
marathon.plot(kind='scatter', x='Age', y='TimeSeconds')
```

```python
is_man = marathon['M/F'] == 'M'
is_woman = marathon['M/F'] == 'F'
```

```python
marathon_m = marathon[is_man]
marathon_w = marathon[is_woman]
```

```python
am = marathon_m.plot(kind='scatter', x='Age', y='TimeSeconds', color='Red')
marathon_w.plot(kind='scatter', x='Age', y='TimeSeconds', ax=am)
```

```python
marathon_m.plot.hexbin(gridsize=20, x='Age', y='TimeSeconds')
```

```python
marathon_w.plot(kind='hexbin', gridsize=20, x='Age', y='TimeSeconds')
```

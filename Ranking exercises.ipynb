```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('../data/Fortune_500_2017.csv',
                      usecols=['Rank','Title','Employees','Profits','Assets'],
                      index_col='Rank')
```

```python
fortune.head(10)
```

```python
fortune.rank()
```

```python
fortune.nlargest(columns='Employees', n=10)
```

```python
fortune['Employees'].rank(ascending=False).head()
```

```python
fortune['RankByEmployee'] = fortune['Employees'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune['Profits'].rank(ascending=False).head()
```

```python
fortune['RankByProfits'] = fortune['Profits'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune.nlargest(n=3, columns='Assets')
```

```python
fortune.nsmallest(n=3, columns='Assets')
```

# data filtering

```python
isEmployeesRankFirst10 = fortune['RankByEmployee'] <= 10
```

```python
isEmployeesRankFirst10.head()
```

```python
isProfitRankFirst10 = fortune.RankByProfits <= 10
```

```python
isProfitRankFirst10.head()
```

```python
fortune[isEmployeesRankFirst10 & isProfitRankFirst10]
```

```python
isEmployeesRankMore400 = fortune.RankByEmployee >= 400
```

```python
fortune[isEmployeesRankMore400 & isProfitRankFirst10]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('../ml_data/high_school_sat_gpa.csv',
                   sep=' ', usecols=['math_SAT','verb_SAT','high_GPA'])
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['math_SAT'], data['high_GPA'], c='red')
plt.show()
```

```python
lr_math = LinearRegression()
```

```python
lr_math.fit(data['math_SAT'].values.reshape(-1,1), data['high_GPA'].values.reshape(-1,1))
```

```python
x_min = data['math_SAT'].min()
```

```python
x_max = data['math_SAT'].max()
```

```python
y_pred = lr_math.predict([[x_min], [x_max]])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['math_SAT'], data['high_GPA'], c='red')
plt.plot([x_min, x_max], y_pred, color='red')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['math_SAT'], data['verb_SAT'], c='red')
plt.show()
```

```python
lr = LinearRegression()
```

```python
lr.fit(data['math_SAT'].values.reshape(-1,1), data['verb_SAT'].values.reshape(-1,1))
```

```python
x_min = data['verb_SAT'].min()
```

```python
x_max = data['verb_SAT'].max()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['math_SAT'], data['verb_SAT'], c='red')
plt.plot([x_min, x_max], lr.predict([[x_min], [x_max]]), color='red')
plt.show()
```

```python
import plotly.express as px
```

```python
df = px.data.tips()
```

```python
df.head()
```

```python
px.scatter(data_frame=df, x='total_bill', y='tip', facet_col='day', category_orders={'day':['Thur', 'Fri', 'Sat','Sun']},
          trendline='ols', color='smoker')
```

```python
px.scatter(data_frame=df, x='total_bill', y='tip', facet_row='time',
          trendline='ols', color='smoker')
```

```python
px.parallel_categories(data_frame=df, color='size')
```
